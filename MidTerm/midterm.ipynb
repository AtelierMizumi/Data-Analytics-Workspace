### Nhan xet

Co the thay cac du lieu nhu square_feet, summary va space thuong de null rat nhieu
Nhung cot nhu bathroom bedrooms, beds co mot so gia tri NULL, co the quy ve gia tri 0

```python
# xu li cac cot co gia tri NULL theo tieu chi da nhan xet

## Loai bo cot square_feet, space va summary

df2 = df2.drop(['square_feet', 'space', 'summary'], axis=1)
df2.sample(7)
```

```python
# Dien du lieu mac dinh cho cac dong bi NULL

## bathrooms, bedrooms, beds se duoc chuyen ve 0.0

df2['bathrooms'] = df2['bathrooms'].fillna(0.0)
df2['bedrooms'] = df2['bedrooms'].fillna(0.0)
df2['beds'] = df2['beds'].fillna(0.0)
df2.isnull().sum()
```

```python
# Loai bo cac dong co gia tri NULL

df2 = df2.dropna()
df2.isnull().sum()
```

```python
# Kiem tra cac outlier

df2.describe()
```

### Co mot nha co tan 15 cai giuong ngu cho 7 phong ngu

```python
# Xu li

mean_value = df2['beds'].mean()

# Thay thế các giá trị lớn nhất trong cột 'beds' bằng giá trị trung bình
max_value = df2['beds'].max()
df2['beds'] = df2['beds'].replace(max_value, mean_value)
```

```python
# Check for duplicates
df2.duplicated().sum()
```

## Du lieu khong co trung lap

```python
# Merge 2 dataset voi nhau

df = pd.merge(df1, df2, how='inner', left_on='listing_id', right_on='id')

# Loai cot id cu

df = df.drop(['id'], axis=1)
df = df.drop(['price_y'], axis=1)
df = df.rename(columns={'price_x': 'price'})
# Loai bo cac price co gia tri NULL
df = df.dropna(subset=['price'])
```

```python
df.sample(7)
```

```python
# Sap xep lai cac dong theo date

df = df.sort_values(by='date')
df.head(7)
```

# Giai quyet cac cau hoi trong bai toan

1. Loai phong nao duoc dat nhieu nhat?
2. Giá trung bình của các danh sách là bao nhiêu? Gia trung binh cua tung loai nha o la bao nhieu?
3. Tỷ lệ phần trăm các phong có sẵn vào ngày 1/4/2016 (hoac bat ky) là bao nhiêu?

```python
# Cau hoi 1: Loai phong duoc dat nhieu nhat
most_booked_room_type = df['room_type'].value_counts().idxmax()
print("Loại phòng được đặt nhiều nhất là:", most_booked_room_type)
```

```python
# Cau hoi 2: Gia trung binh cua tung loai nha o la bao nhieu
# Convert the 'price' column to a numeric type
df['price'] = df['price'].replace('').astype(float)
average_price = df['price'].mean()
average_price_per_room_type = df.groupby('room_type')['price'].mean()
print("Giá trung bình của các danh sách là:", average_price)
print("Giá trung bình của từng loại nhà ở là:")
print(average_price_per_room_type)
```

Vay thue nha nguyen can/ can ho la dat nhat
Thue phong o ghep la re nhat

```python
if 'date' in df.columns and df['date'].dtype != 'datetime64[ns]':
    df['date'] = pd.to_datetime(df['date'])
sample_date = df['date'].unique()[0]  # Just using the first available date as an example
available_listings = df[df['date'] == sample_date]['available'].value_counts(normalize=True) * 100
print(f"Tỷ lệ phần trăm các phòng có sẵn vào ngày {sample_date} là:")
print(available_listings)
```

```python
# Mot chut visualisation
import matplotlib.pyplot as plt

# Đồ thị cột cho giá trung bình của từng loại nhà ở
average_price_per_room_type.plot(kind='bar', color='skyblue')
plt.title('Giá trung bình của từng loại nhà ở')
plt.show()
```
