# Thuc hanh Visualize du lieu

Trong tiet hoc nay (von la thi giua ki) chung ta se hoc cac trinh bay du lieu

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Nhap du lieu vao
df = pd.read_csv('Ice Cream Ratings.csv')
print(df.info())
print(df.head())
```

```python
df.plot.pie(y='Overall Rating', figsize=(5, 5))
```

```python
df.plot()
```

```python
df.plot(x="Date", y="Overall Rating", kind="bar")
plt.title('Ice Cream Ratings')
```

```python
df.plot(kind='scatter', x='Date', y='Overall Rating', color='red', s=500)
```

```python
df.plot(kind='bar')
```

```python

# Display side way
df.plot(kind='barh', title="Ice Cream")
```
