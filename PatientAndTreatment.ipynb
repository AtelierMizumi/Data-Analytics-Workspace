```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options for better visibility
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 20)

# Load the datasets
patients = pd.read_csv('patients.csv')
treatments = pd.read_csv('treatments.csv')

print(patients.head())
print(treatments.head())
```

```python
# Analyzing patients dataset
print("\n======== PATIENTS DATASET ========\n")

print("Quick look at the data (first 5 rows):")
print(patients.head())

print("\nQuick look at the data (last 5 rows):")
print(patients.tail())

print("\nRandom sample (3 rows):")
print(patients.sample(3))

print("\nData shape (rows, columns):")
print(patients.shape)

print("\nColumn names:")
print(patients.columns.tolist())

print("\nData information:")
print(patients.info())

print("\nDescriptive statistics:")
print(patients.describe(include='all').T)

print("\nMissing values count by column:")
print(patients.isna().sum())

print("\nMissing values percentage by column:")
print((patients.isna().sum() / len(patients) * 100).round(2))

print("\nDuplicates count:")
print(patients.duplicated().sum())

# For each column, check unique values
print("\nUnique values for categorical columns:")
for col in patients.select_dtypes(include=['object']).columns:
    n_unique = patients[col].nunique()
    if n_unique < 20:  # Only show if not too many unique values
        print(f"\n{col} ({n_unique} unique values):")
        print(patients[col].value_counts())
    else:
        print(f"\n{col}: {n_unique} unique values")
```

```python

# Repeat analysis for treatments dataset
print("\n\n======== TREATMENTS DATASET ========\n")

print("Quick look at the data (first 5 rows):")
print(treatments.head())

print("\nQuick look at the data (last 5 rows):")
print(treatments.tail())

print("\nRandom sample (3 rows):")
print(treatments.sample(3))

print("\nData shape (rows, columns):")
print(treatments.shape)

print("\nColumn names:")
print(treatments.columns.tolist())

print("\nData information:")
print(treatments.info())

print("\nDescriptive statistics:")
print(treatments.describe(include='all').T)

print("\nMissing values count by column:")
print(treatments.isna().sum())

print("\nMissing values percentage by column:")
print((treatments.isna().sum() / len(treatments) * 100).round(2))

print("\nDuplicates count:")
print(treatments.duplicated().sum())

# For each column, check unique values
print("\nUnique values for categorical columns:")
for col in treatments.select_dtypes(include=['object']).columns:
    n_unique = treatments[col].nunique()
    if n_unique < 20:  # Only show if not too many unique values
        print(f"\n{col} ({n_unique} unique values):")
        print(treatments[col].value_counts())
    else:
        print(f"\n{col}: {n_unique} unique values")
```

```python
# Optional: Visualize missing data
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.heatmap(patients.isna(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Values in Patients Dataset')

plt.subplot(1, 2, 2)
sns.heatmap(treatments.isna(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Values in Treatments Dataset')
plt.tight_layout()
plt.show()
```
